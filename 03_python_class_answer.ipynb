## ■(1) 関数定義
### ●問題
2つの数値を引数として渡すと、その値を乗算した結果を返す「乗算関数」(mul)を定義し、実行してください。
<br>
### ●実行結果(例:引数に123と456を渡した場合)
56088<br>


```python
def mul(n1, n2):
  return n1 * n2

print(mul(123, 456))
```

## ■(2) クラスとメソッドの定義
### ●問題<br>
2つの引数を渡すと乗算した結果を返す「乗算メソッド」(mul)を含む「計算クラス」(Calc)を定義して、実行してください。
<br>
### ●実行結果<br>
※定義のみのため表示はありません。実行のみをして下さい。

```python
class Calc:
  def mul(self, n1, n2):
    return n1 * n2
```

## ■(3) インスタンスの生成
### ●問題<br>
(2)で定義した「計算クラス」(Calc)のインスタンスを生成し、引数に456と789を渡して「乗算メソッド」(mul)を呼び出してください。<br>
### ●実行結果<br>

```python
calc = Calc()
print(calc.mul(456, 789))
```

## ■(4) コンストラクタ
### ●問題
「動物クラス」(Animal)を定義して、引数に文字列を渡すと「名前」のインスタンス変数(name)に代入するコンストラクタを定義してください。さらに、名前表示のメソッド(disp)を定義して、インスタンスを生成し、そのメソッドを呼び出してください。<br>
### ●実行結果(例：名前に「どうぶつ」を渡したケース)
私の名前は どうぶつ です。<br>

```python
class Animal:
  def __init__(self, name): # コンストラクタ(初期化関数)
    self.name = name
  def disp(self): # 名前表示メソッド
    print('私の名前は',self.name, 'です。')

a = Animal('どうぶつ')
a.disp() # 名前表示メソッド
```

## ■(5) 継承
### ●問題<br>
(4)で定義した「動物クラス」(Animal)を継承し、「猫クラス」(Cat)を定義してください。猫クラスには、呼び出すと「にゃー！」と表示する「鳴く」(cries)メソッドを定義してください。最後に、猫クラスのインスタンスを生成し、名前表示メソッドと鳴くメソッドを呼び出してください。
<br>
### ●実行結果(例：インスタンス生成時に「タマ」を渡したケース)<br>
私の名前は タマ です。<br>
にゃー！<br>

```python
class Cat(Animal):
  def __init__(self, name):
    Animal.__init__(self, name) # 親クラスのコンストラクタを呼び出す
  def cries(self): # 新しく追加した鳴くメソッド
    print('にゃー！')

c = Cat('タマ')
c.disp()  # 親クラスの名前表示メソッド
c.cries() # 新しく追加した鳴くメソッド
```

## ■(6) オーバーライド
### ●問題<br>
(5)で定義した猫クラスを継承し、犬クラス(Dog)を定義してください。さらに、鳴く(cries)メソッドと同じ名前のメソッドを再定義(オーバーライド)し、「わん！」という文字列を表示してください。最後に、犬クラスのインスタンスを生成し、名前表示メソッドと、鳴くメソッド呼び出してください。
<br>
### ●実行結果(インスタンス生成時に「ポチ」を渡したケース)<br>
私の名前は ポチ です。<br>
わん！<br>

```python
class Dog(Cat):
  def __init__(self, name):
    Animal.__init__(self, name) # 親クラスのコンストラクタを呼び出す
  def cries(self): # 再定義した鳴くメソッド
    print('わん！')

d = Dog('ポチ')
d.disp()  # 親クラスの名前表示メソッド
d.cries() # 再定義した鳴くメソッド
```

## ■(7) 多重継承(オーブンレンジ)
### ●問題
まずは下記のコードを実行し、オーブン(Oven)とレンジ(Range)のクラスを定義してください。次のボックスに、オーブンとレンジを多重継承したオーブンレンジ(Ovenrange)クラスを定義してください。ここではクラス名のみを定義します。最後に、オーブンレンジクラスのインスタンスを生成し、焼く(bake)と温める(warm)のメソッドを呼び出してください。<br>
### ●実行結果
焼きます！<br>
温めます！<br>

```python
# このコードを実行してください
class Oven:
  def bake(self): # 焼くメソッド
    print('焼きます！')
class Range:
  def warm(self): # 温めるメソッド
    print('温めます！')
```

```python
class Ovenrange(Oven, Range):
  pass

o = Ovenrange()
o.bake()
o.warm()
```

## ■(8) 総合演習(銀行クラス)
次の仕様に従って銀行クラスを定義し、残高照会、口座入金、口座出金のメソッドを呼び出してください。<br>
### ●仕様
・クラス名：Bank<br>
・インスタンス変数：account<br>
・コンストラクタ：引数の値をaccountに代入する<br>
・口座入金(add)：引数の値をaccountに足して、引数の値を返す<br>
・口座出金(sub)：引数の値をaccountから引いて、引数の値を返す<br>
・残高照会(balance)：accountの値を返す<br>
### ●実行結果(インスタンス生成時に1000を渡したケース)
残高照会: 1000<br>
口座入金: 5000<br>
残高照会: 6000<br>
口座出金: 2000<br>
残高照会: 4000<br>





```python
# 銀行クラスの定義
class Bank:
  def __init__(self, n):
    self.account = n
  def add(self, n):
    self.account += n
    return n
  def sub(self, n):
    self.account -= n
    return n
  def balance(self):
    return self.account

# インスタンスの生成
bk = Bank(1000) # 口座開設で入金:1000
print('残高照会:', bk.balance())
print('口座入金:', bk.add(5000)) # 入金:5000
print('残高照会:', bk.balance())
print('口座出金:', bk.sub(2000)) # 出金:2000
print('残高照会:', bk.balance())
```
