## ■(1) 関数定義
### ●問題
2つの数値を引数として渡すと、その値を加算した結果を返す「加算関数」(add)を定義し、実行してください。
<br>
### ●実行結果(例：引数に123と456を渡した場合)
579<br>


```python
def add(n1, n2):
  return n1 + n2

print(add(123, 456))
```

## ■(2) クラスとメソッドの定義
### ●問題<br>
2つの引数を渡すと加算した結果を返す「加算メソッド」(add)を含む「計算クラス」(Calc)を定義して、実行してください。
<br>
### ●実行結果<br>
※定義のみのため表示はありません。実行のみをして下さい。

```python
class Calc:
  def add(self, n1, n2):
    return n1 + n2
```

## ■(3) インスタンスの生成
### ●問題<br>
(2)で定義した「計算クラス」(Calc)のインスタンスを生成し、引数に456と789を渡して「加算メソッド」(add)を呼び出してください。<br>
### ●実行結果<br>

```python
calc = Calc()
print(calc.add(456, 789))
```

## ■(4) コンストラクタ
### ●問題<br>
「鳥類クラス」(Bird)を定義して、引数に文字列を渡すと「名前」のインスタンス変数(name)に代入するコンストラクタを定義してください。さらに、名前表示のメソッド(disp)を定義して、インスタンスを生成し、そのメソッドを呼び出してください。
<br>
### ●実行結果(例：名前に「ひよこ」を渡したケース)<br>
私の名前は ひよこ です。

```python
class Bird:
  def __init__(self, name): # コンストラクタ(初期化関数)
    self.name = name
  def disp(self): # 名前表示メソッド
    print('私の名前は', self.name, 'です。')

b = Bird('ひよこ')
b.disp() # 名前表示メソッド
```

## ■(5) 継承
### ●問題
(4)で定義した「鳥類クラス」(Bird)を継承し、「鶏クラス」(Chicken)を定義してください。鶏クラスには、呼び出すと「コケコッコー！」と表示する「鳴く」(cries)メソッドを定義してください。最後に、鶏クラスのインスタンスを生成し、名前表示メソッドと鳴くメソッドを呼び出してください。
<br>
### ●実行結果(例：インスタンス生成時に「チキン」を渡したケース)
私の名前は チキン です<br>
コケコッコー！<br>

```python
class Chicken(Bird):
  def __init__(self, name):
    Bird.__init__(self, name) # 親クラスのコンストラクタを呼び出す
  def cries(self): # 新しく追加した鳴くメソッド
    print('コケコッコー！')

c = Chicken('チキン')
c.disp()  # 親クラスの名前表示メソッド
c.cries() # 新しく追加した鳴くメソッド
```

## ■(6) オーバーライド
### ●問題
(5)で定義した「鶏クラス」(Chicken)を継承し、「カラスクラス」(Crow)を定義してください。さらに、「鳴く」(cries)メソッドと同じ名前のメソッドを再定義(オーバーライド)し、「カーカー！」という文字列を表示してください。最後に、カラスクラスのインスタンスを生成し、名前表示メソッドと、鳴くメソッド呼び出してください。<br>
### ●実行結果(例：インスタンス生成時に「カラス」を渡したケース)
私の名前は カラス です。<br>
カーカー！<br>

```python
class Crow(Chicken):
  def __init__(self, name):
    Chicken.__init__(self, name) # 親クラスのコンストラクタを呼び出す
  def cries(self): # 再定義した鳴くメソッド
    print('カーカー！')

c = Crow('カラス')
c.disp()  # 親クラスの名前表示メソッド
c.cries() # 再定義した鳴くメソッド
```

## ■(7) 多重継承
### ●問題
まずは下記のコードを実行し、ラジオ(Radio)とカセット(Cassette)のクラスを定義してください。次のボックスに、ラジオとカセットを多重継承したラジカセ(Radicas)クラスを定義してください。ここではクラス名のみを定義します。最後に、ラジカセクラスのインスタンスを生成し、受信(receive)と再生(play)のメソッドを呼び出してください。<br>
### ●実行結果
ラジオを受信します！<br>
テープを再生します！<br>

```python
# このコードを実行してください
class Radio:
  def receive(self): # 受信メソッド
    print('ラジオを受信します！')

class Cassette: # 再生メソッド
  def play(self):
    print('テープを再生します！')
```

```python
class Radicas(Radio, Cassette):
  pass # 何も定義しないときに記述する

r = Radicas()
r.receive() # 受信メソッド
r.play()    # 再生メソッド
```
